### **Bibliotecas**

Importa todas as bibliotecas necessárias.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### **Importação dos dados**



Os dados de cada loja estão disponíveis nos arquivos CSV e serão carregados utilizando o código base já fornecido. O código extrai informações das quatro lojas e organiza os dados em DataFrames utilizando a biblioteca **Pandas**.

```python
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

#**1. Análise do faturamento**


Calcula o **faturamento total** de cada loja. Somando os valores da coluna **Preço** de cada loja para estimar o faturamento.

```python
faturamentoLoja1 = loja1['Preço'].sum()
faturamentoLoja2 = loja2['Preço'].sum()
faturamentoLoja3 = loja3['Preço'].sum()
faturamentoLoja4 = loja4['Preço'].sum()

print('Faturamento Loja 1:', 'R$ {:,.2f}'.format(faturamentoLoja1))
print('Faturamento Loja 2:', 'R$ {:,.2f}'.format(faturamentoLoja2))
print('Faturamento Loja 3:', 'R$ {:,.2f}'.format(faturamentoLoja3))
print('Faturamento Loja 4:', 'R$ {:,.2f}'.format(faturamentoLoja4))
```

```python
relacaoFaturamento = {'Loja 1': faturamentoLoja1, 'Loja 2': faturamentoLoja2, 'Loja 3': faturamentoLoja3, 'Loja 4': faturamentoLoja4}

plt.pie(relacaoFaturamento.values(), labels=relacaoFaturamento.keys(), autopct='%1.1f%%')
plt.title('Relação de Faturamento')
plt.axis('equal')
plt.show()
```

# **2. Vendas por Categoria**


Calcular a **quantidade de produtos vendidos** por **categoria** em cada loja. A ideia é agrupar os dados por categoria e contar o número de vendas de cada tipo, mostrando **as categoria mais populares** de cada loja.

```python
categoriasLoja1 = loja1['Categoria do Produto'].value_counts()
categoriasLoja2 = loja2['Categoria do Produto'].value_counts()
categoriasLoja3 = loja3['Categoria do Produto'].value_counts()
categoriasLoja4 = loja4['Categoria do Produto'].value_counts()

print("Categorias mais vendidas - Loja 1:")
print(categoriasLoja1.to_string())
print("\nCategorias mais vendidas - Loja 2:")
print(categoriasLoja2.to_string())
print("\nCategorias mais vendidas - Loja 3:")
print(categoriasLoja3.to_string())
print("\nCategorias mais vendidas - Loja 4:")
print(categoriasLoja4.to_string())
```

```python
categoriasCombinadas =
pd.concat([categoriasLoja1, categoriasLoja2, categoriasLoja3, categoriasLoja4],
          axis=1, keys=['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'])
categoriasCombinadas.plot(kind='bar', figsize=(12, 6))
plt.title('Vendas por Categoria em Todas as Lojas')
plt.xlabel('Categoria do Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# **3. Média de Avaliação das Lojas**

Calcula a média das avaliações dos clientes para cada loja. O objetivo é **entender a satisfação dos clientes com os produtos vendidos**.

```python
mediaAvaliacaoLoja1 = loja1['Avaliação da compra'].mean()
mediaAvaliacaoLoja2 = loja2['Avaliação da compra'].mean()
mediaAvaliacaoLoja3 = loja3['Avaliação da compra'].mean()
mediaAvaliacaoLoja4 = loja4['Avaliação da compra'].mean()

print(f'Média de Avaliação \n'
      f'Loja 1: {mediaAvaliacaoLoja1:.2f} \n'
      f'Loja 2: {mediaAvaliacaoLoja2:.2f} \n'
      f'Loja 3: {mediaAvaliacaoLoja3:.2f} \n'
      f'Loja 4: {mediaAvaliacaoLoja4:.2f} \n')
```

```python
mediasAvaliacao = {
    'Loja 1': mediaAvaliacaoLoja1,
    'Loja 2': mediaAvaliacaoLoja2,
    'Loja 3': mediaAvaliacaoLoja3,
    'Loja 4': mediaAvaliacaoLoja4
}

plt.scatter(mediasAvaliacao.keys(), mediasAvaliacao.values())
plt.title('Média de Avaliação das Lojas')
plt.xlabel('Lojas')
plt.ylabel('Média de Avaliação')
plt.grid(True)
plt.show()
```

# **4. Produtos Mais e Menos Vendidos**

Identifica quais produtos foram os mais vendidos e os menos vendidos em cada loja.  Ao final, visualize os resultados de forma que fique claro quais produtos se destacaram nas vendas de cada loja.

```python
qtdVendidaLoja1 = loja1.groupby('Produto').size().sort_values(ascending=False)
qtdVendidaLoja2 = loja2.groupby('Produto').size().sort_values(ascending=False)
qtdVendidaLoja3 = loja3.groupby('Produto').size().sort_values(ascending=False)
qtdVendidaLoja4 = loja4.groupby('Produto').size().sort_values(ascending=False)

maiorLoja1 = qtdVendidaLoja1.head(5)
menorLoja1 = qtdVendidaLoja1.tail(5)

maiorLoja2 = qtdVendidaLoja2.head(5)
menorLoja2 = qtdVendidaLoja2.tail(5)

maiorLoja3 = qtdVendidaLoja3.head(5)
menorLoja3 = qtdVendidaLoja3.tail(5)

maiorLoja4 = qtdVendidaLoja4.head(5)
menorLoja4 = qtdVendidaLoja4.tail(5)

print("Produtos mais vendidos - Loja 1:")
print(maiorLoja1.to_string())
print("\nProdutos mais vendidos - Loja 2:")
print(maiorLoja2.to_string())
print("\nProdutos mais vendidos - Loja 3:")
print(maiorLoja3.to_string())
print("\nProdutos mais vendidos - Loja 4:")
print(maiorLoja4.to_string())

print("\nProdutos menos vendidos - Loja 1:")
print(menorLoja1.to_string())
print("\nProdutos menos vendidos - Loja 2:")
print(menorLoja1.to_string())
print("\nProdutos menos vendidos - Loja 3:")
print(menorLoja1.to_string())
print("\nProdutos menos vendidos - Loja 4:")
print(menorLoja1.to_string())
```

```python
productosLoja1 = pd.concat([maiorLoja1, menorLoja1])
productosLoja2 = pd.concat([maiorLoja2, menorLoja2])
productosLoja3 = pd.concat([maiorLoja3, menorLoja3])
productosLoja4 = pd.concat([maiorLoja4, menorLoja4])

plt.figure(figsize=(10, 6))
colors = ['blue'] * len(maiorLoja1) + ['red'] * len(menorLoja1)
plt.bar(productosLoja1.index, productosLoja1.values, color=colors)
plt.title('Quantidade Vendida dos Produtos Mais e Menos Vendidos - Loja 1')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(productosLoja2.index, productosLoja2.values, color=colors)
plt.title('Quantidade Vendida dos Produtos Mais e Menos Vendidos - Loja 2')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(productosLoja3.index, productosLoja3.values, color=colors)
plt.title('Quantidade Vendida dos Produtos Mais e Menos Vendidos - Loja 3')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(productosLoja4.index, productosLoja4.values, color=colors)
plt.title('Quantidade Vendida dos Produtos Mais e Menos Vendidos - Loja 4')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# **5. Frete Médio por Loja**


Calcular o custo médio de frete para cada loja. O objetivo é entender quanto, em média, está sendo gasto com frete para cada uma das lojas.

```python
freteMedioLoja1 = loja1['Frete'].mean()
freteMedioLoja2 = loja2['Frete'].mean()
freteMedioLoja3 = loja3['Frete'].mean()
freteMedioLoja4 = loja4['Frete'].mean()

print('Frete médio 1:', 'R$ {:,.2f}'.format(freteMedioLoja1))
print('Frete médio 2:', 'R$ {:,.2f}'.format(freteMedioLoja2))
print('Frete médio 3:', 'R$ {:,.2f}'.format(freteMedioLoja3))
print('Frete médio 4:', 'R$ {:,.2f}'.format(freteMedioLoja4))
```

```python
freteMedio = {
    'Loja 1': freteMedioLoja1,
    'Loja 2': freteMedioLoja2,
    'Loja 3': freteMedioLoja3,
    'Loja 4': freteMedioLoja4
}

plt.figure(figsize=(8, 5))
plt.bar(freteMedio.keys(), freteMedio.values(), color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Frete Médio por Loja')
plt.xlabel('Lojas')
plt.ylabel('Frete Médio (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
